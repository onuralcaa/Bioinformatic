```python
import os
from Bio import SeqIO

# Klasör isimlerini tanımlayın
folders = ["DNA_sequence_10", "Protein_sequence_10"]

# Her klasördeki dosyaları listeleyip okuyalım
for folder in folders:
    print("\n\n\n")
    print("=" * 50)
    print(f"Klasör: {folder}")
    print("=" * 50)
    
    
    folder_path = os.path.join(os.getcwd(), folder)  # Klasör yolunu oluştur
    files = os.listdir(folder_path)  # Klasördeki dosyaları listele
    
    # Her dosyayı ayrı bir başlık altında göster
    for file in files:
        print(f"\n--- Dosya: {file} ---")
        
        file_path = os.path.join(folder_path, file)  # Dosyanın tam yolunu al
        
        # Dosya içeriğini okuma
        for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
            print(f"ID: {record.id}")
            print(f"Dizi: {record.seq}")
            print(f"Dizi Uzunluğu: {len(record.seq)}")
```
