```python
from Bio.SeqRecord import SeqRecord

# DNA sekans klasörü
input_folder = "DNA_sequence_10"

# Yeni protein dosyalarının kaydedileceği klasör
output_folder = "Protein_sequences_from_DNA"
os.makedirs(output_folder, exist_ok=True)  # Klasörü oluştur (varsa hata vermez)

# Klasör yolunu oluştur
input_folder_path = os.path.join(os.getcwd(), input_folder)
files = os.listdir(input_folder_path)  # Klasördeki dosyaları listele

# Her dosyayı okuyup proteine çevirme ve kaydetme
for file in files:
    print(f"\n--- Dosya: {file} ---")
    file_path = os.path.join(input_folder_path, file)  # Dosyanın tam yolunu al
    
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        dna_sequence = str(record.seq)  # DNA dizisini stringe çevir
        
        # Proteine çevirme
        protein_sequence = Seq(dna_sequence).translate(to_stop=True)  # Stop kodonlarında dur
        
        # Yeni bir SeqRecord oluştur
        protein_record = SeqRecord(
            protein_sequence,  # Protein dizisi
            id=record.id,  # Orijinal DNA sekansının ID'sini kullan
            description="Protein sequence translated from DNA"
        )
        
        # Protein dosyasını kaydetme
        output_file_path = os.path.join(output_folder, f"{record.id}_protein.fasta")
        with open(output_file_path, "w") as output_file:
            SeqIO.write(protein_record, output_file, "fasta")
        
        print(f"Protein dizisi {output_file_path} dosyasına kaydedildi.")
```

## E-) DNA sıralarının GC içeriğinin hesaplanması

```python
# DNA sekans klasörü
input_folder = "DNA_sequence_10"

# Klasör yolunu oluştur
input_folder_path = os.path.join(os.getcwd(), input_folder)
files = os.listdir(input_folder_path)  # Klasördeki dosyaları listele

# GC içeriğini hesaplayan fonksiyon
def calculate_gc_content(sequence):
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    gc_content = ((g_count + c_count) / len(sequence)) * 100
    return gc_content

# Her dosya için GC içeriğini hesaplama
for file in files:
    print(f"\n--- Dosya: {file} ---")
    file_path = os.path.join(input_folder_path, file)  # Dosyanın tam yolunu al
    
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        dna_sequence = str(record.seq)  # DNA dizisini stringe çevir
        gc_content = calculate_gc_content(dna_sequence)  # GC içeriğini hesapla
        
        print(f"ID: {record.id}")
        print(f"DNA Dizi Uzunluğu: {len(dna_sequence)}")
        print(f"GC İçeriği: {gc_content:.2f}%")
        print(f"\n*****************************************************************************************************")

```

## H-) Pozisyon Sıklık Matrislerinin (PFM) bulunması

```python
# Klasör adı
folder = "DNA_sequence_10"
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]
import numpy as np

# Nükleotidler
nucleotides = ['A', 'T', 'C', 'G']

# Her bir dosyayı işleyip PFM hesaplama
for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequence = str(record.seq).upper()  # Sekansı büyük harfe çeviriyoruz
        
        # DNA sırasını 35 bazlık parçalar halinde bölelim
        chunk_size = 10
        sequences = [sequence[i:i + chunk_size] for i in range(0, len(sequence), chunk_size)]
        
        # Maksimum uzunluk belirleme
        max_length = max(len(seq) for seq in sequences)
        
        # Sekansları hizalama
        aligned_sequences = [seq.ljust(max_length, '-') for seq in sequences]
        
        # Pozisyon Frekans Matrisi (PFM) hesaplama
        pfm = np.zeros((len(nucleotides), max_length), dtype=int)
        
        for i, nucleotide in enumerate(nucleotides):
            for j in range(max_length):
                pfm[i, j] = sum(1 for seq in aligned_sequences if seq[j] == nucleotide)
        
        # PFM'yi formatlı şekilde yazdırma
        print(f"\n=== DNA Sırası: {record.id} ===")
        print("Pozisyon Frekans Matrisi (PFM):")
        print("          " + "  ".join(nucleotides))
        for pos, values in enumerate(pfm.T):
            print(f"{pos:9d} " + " ".join(f"{val:2d}" for val in values))
```

## I-) Pozisyon Olasılık Matrislerinin (PPM) bulunması

```python
# Klasör adı
folder = "DNA_sequence_10"
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]

# Nükleotidler
nucleotides = ['A', 'T', 'C', 'G']

# Her bir dosyayı işleyip PPM hesaplama
for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequence = str(record.seq).upper()  # Sekansı büyük harfe çeviriyoruz
        
        # DNA sırasını 10 bazlık parçalar halinde bölelim
        chunk_size = 10
        sequences = [sequence[i:i + chunk_size] for i in range(0, len(sequence), chunk_size)]
        
        # Maksimum uzunluk belirleme
        max_length = max(len(seq) for seq in sequences)
        
        # Sekansları hizalama
        aligned_sequences = [seq.ljust(max_length, '-') for seq in sequences]
        
        # Pozisyon Frekans Matrisi (PFM) hesaplama
        pfm = np.zeros((len(nucleotides), max_length), dtype=int)
        
        for i, nucleotide in enumerate(nucleotides):
            for j in range(max_length):
                pfm[i, j] = sum(1 for seq in aligned_sequences if seq[j] == nucleotide)
        
        # Pozisyon Olasılık Matrisi (PPM) hesaplama
        sequence_count = len(aligned_sequences)
        ppm = pfm / sequence_count
        
        # Sadece PPM'yi formatlı şekilde yazdırma
        print(f"\n=== DNA Sırası: {record.id} ===")
        print("Pozisyon Olasılık Matrisi (PPM):")
        print("          " + "  ".join(nucleotides))
        for pos, values in enumerate(ppm.T):
            print(f"{pos:9d} " + " ".join(f"{val:.2f}" for val in values))
```

## J-) Pozisyon Ağırlık Matrisinin (PWM) bulunması

```python
# Klasör adı
folder = "DNA_sequence_10"
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]

# Nükleotidler
nucleotides = ['A', 'T', 'C', 'G']

# Her bir dosyayı işleyip PWM hesaplama
for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequence = str(record.seq).upper()  # Sekansı büyük harfe çeviriyoruz
        
        # DNA sırasını 10 bazlık parçalar halinde bölelim
        chunk_size = 10
        sequences = [sequence[i:i + chunk_size] for i in range(0, len(sequence), chunk_size)]
        
        # Maksimum uzunluk belirleme
        max_length = max(len(seq) for seq in sequences)
        
        # Sekansları hizalama
        aligned_sequences = [seq.ljust(max_length, '-') for seq in sequences]
        
        # Pozisyon Frekans Matrisi (PFM) hesaplama
        pfm = np.zeros((len(nucleotides), max_length), dtype=int)
        
        for i, nucleotide in enumerate(nucleotides):
            for j in range(max_length):
                pfm[i, j] = sum(1 for seq in aligned_sequences if seq[j] == nucleotide)
        
        # Pozisyon Olasılık Matrisi (PPM) hesaplama
        sequence_count = len(aligned_sequences)
        ppm = pfm / sequence_count
        
        # Pozisyon Ağırlık Matrisi (PWM) hesaplama
        background_prob = 0.25  # Nükleotidler için varsayılan arka plan olasılığı
        pwm = np.log2(ppm / background_prob, where=(ppm > 0))  # Log2(PPM / background_prob)
        pwm[np.isnan(pwm)] = 0  # NaN olan yerleri 0 olarak doldur
        
        # Sadece PWM'yi formatlı şekilde yazdırma
        print(f"\n=== DNA Sırası: {record.id} ===")
        print("Pozisyon Ağırlık Matrisi (PWM):")
        print("          " + "  ".join(nucleotides))
        for pos, values in enumerate(pwm.T):
            print(f"{pos:9d} " + " ".join(f"{val:.2f}" for val in values))
```

```python
import os
from Bio import SeqIO
from Bio.Align import PairwiseAligner

# Klasör adı
folder = "Protein_sequence_10"

# Klasördeki dosyaları listele
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]

# Protein sıralarını yükle
sequences = []
for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequences.append((record.id, str(record.seq)))

# PairwiseAligner oluştur
aligner = PairwiseAligner()
aligner.mode = "global"
aligner.match_score = 1.0  # Eşleşme skoru
aligner.mismatch_score = -1.0  # Uyuşmazlık skoru
aligner.gap_score = -2.0  # Boşluk (gap) skoru

# Global hizalamaları hesapla
print("\nProtein Sıralarının Global Hizalama Sonuçları:\n" + "=" * 50)
for i in range(len(sequences)):
    for j in range(i + 1, len(sequences)):
        id1, seq1 = sequences[i]
        id2, seq2 = sequences[j]
        
        # Hizalamayı hesapla ve sadece en iyi sonucu al
        alignment = aligner.align(seq1, seq2)[0]  # Sadece en yüksek skoru olan hizalamayı seç
        
        print(f"\n--- Hizalanan Sıralar: {id1} ve {id2} ---")
        print(f"Hizalama Skoru: {alignment.score}")
        print(alignment)  # Sadece en iyi hizalamayı yazdır
        print("\n" + "*" * 80)
```

## D-) B ve C şıkları için farklı match, mismatch ve GAP skorlarına göre yeniden hizalama

### 3 farklı kombinasyon ile hizalama yapılmıştır:
#### match = +2, mismatch = -1, GAP = -3
#### match = +3 mismatch = -2, GAP = -2
#### match = +4, mismatch = -3, GAP = -1

```python
# Klasör adı
folder = "Protein_sequence_10"

# Klasördeki dosyaları listele
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]

# Protein sıralarını yükle
sequences = []
for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequences.append((record.id, str(record.seq)))

# Skor parametreleri
scoring_schemes = [
    {"match": 2.0, "mismatch": -1.0, "gap": -3.0},
    {"match": 3.0, "mismatch": -2.0, "gap": -2.0},
    {"match": 4.0, "mismatch": -3.0, "gap": -1.0},
]

# Hizalama sonuçlarını hesapla
for scheme in scoring_schemes:
    print(f"\n=== Match: {scheme['match']}, Mismatch: {scheme['mismatch']}, Gap: {scheme['gap']} ===\n")

    # Global ve Lokal hizalama için PairwiseAligner oluştur
    global_aligner = Align.PairwiseAligner()
    global_aligner.mode = "global"
    global_aligner.match_score = scheme["match"]
    global_aligner.mismatch_score = scheme["mismatch"]
    global_aligner.gap_score = scheme["gap"]

    local_aligner = Align.PairwiseAligner()
    local_aligner.mode = "local"
    local_aligner.match_score = scheme["match"]
    local_aligner.mismatch_score = scheme["mismatch"]
    local_aligner.gap_score = scheme["gap"]

    # Her sıralama çifti için hizalama
    for i in range(len(sequences)):
        for j in range(i + 1, len(sequences)):
            id1, seq1 = sequences[i]
            id2, seq2 = sequences[j]
            
            # Global hizalama
            global_alignment = global_aligner.align(seq1, seq2)
            print(f"\n--- Global Hizalama: {id1} ve {id2} ---")
            print(f"Hizalama Skoru (Global): {global_alignment.score}")
            print(global_alignment[0])
            
            # Lokal hizalama
            local_alignment = local_aligner.align(seq1, seq2)
            print(f"\n--- Lokal Hizalama: {id1} ve {id2} ---")
            print(f"Hizalama Skoru (Lokal): {local_alignment.score}")
            print(local_alignment[0])
```
