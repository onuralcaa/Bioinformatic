## A-) DNA Nükleotid sayılarının yazdırılması

```python
# DNA sekans klasörü
folder = "DNA_sequence_10"

# Klasör yolunu oluştur
folder_path = os.path.join(os.getcwd(), folder)
files = os.listdir(folder_path)  # Klasördeki dosyaları listele

# Her dosyayı okuyup nükleotid sayılarını hesapla
for file in files:
    print(f"\n--- Dosya: {file} ---")
    file_path = os.path.join(folder_path, file)  # Dosyanın tam yolunu al
    
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        sequence = str(record.seq)  # Sekansı stringe çevir
        nucleotide_counts = {
            "A": sequence.count("A"),
            "T": sequence.count("T"),
            "C": sequence.count("C"),
            "G": sequence.count("G")
        }
        
        # Sonuçları yazdır
        print(f"ID: {record.id}")
        print(f"Dizi Uzunluğu: {len(sequence)}")
        print(f"Nükleotid Sayıları: {nucleotide_counts}")
```

## C-) Sıraların Protein' e çevrilmesi (kodonlar kullanılarak)

```python
from Bio.Seq import Seq

# DNA sekans klasörü
folder = "DNA_sequence_10"

# Klasör yolunu oluştur
folder_path = os.path.join(os.getcwd(), folder)
files = os.listdir(folder_path)  # Klasördeki dosyaları listele

# Her dosyayı okuyup proteine çevirme
for file in files:
    print(f"\n--- Dosya: {file} ---")
    file_path = os.path.join(folder_path, file)  # Dosyanın tam yolunu al
    
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        dna_sequence = str(record.seq)  # DNA dizisini stringe çevir
        
        # Proteine çevirme
        protein_sequence = Seq(dna_sequence).translate(to_stop=True)  # Stop kodonlarında dur
        
        # Sonuçları yazdır
        print(f"ID: {record.id}")
        print(f"DNA Dizisi: {dna_sequence}")
        print(f"\nProtein Dizisi: {protein_sequence}")
        print(f"\n*****************************************************************************************************")
```

## D-) C adımında bulunan Protein sıralarının fasta formatında farklı bir dosyaya kaydedilmesi

```python
from Bio.SeqRecord import SeqRecord

# DNA sekans klasörü
input_folder = "DNA_sequence_10"

# Yeni protein dosyalarının kaydedileceği klasör
output_folder = "Protein_sequences_from_DNA"
os.makedirs(output_folder, exist_ok=True)  # Klasörü oluştur (varsa hata vermez)

# Klasör yolunu oluştur
input_folder_path = os.path.join(os.getcwd(), input_folder)
files = os.listdir(input_folder_path)  # Klasördeki dosyaları listele

# Her dosyayı okuyup proteine çevirme ve kaydetme
for file in files:
    print(f"\n--- Dosya: {file} ---")
    file_path = os.path.join(input_folder_path, file)  # Dosyanın tam yolunu al
    
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        dna_sequence = str(record.seq)  # DNA dizisini stringe çevir
        
        # Proteine çevirme
        protein_sequence = Seq(dna_sequence).translate(to_stop=True)  # Stop kodonlarında dur
        
        # Yeni bir SeqRecord oluştur
        protein_record = SeqRecord(
            protein_sequence,  # Protein dizisi
            id=record.id,  # Orijinal DNA sekansının ID'sini kullan
            description="Protein sequence translated from DNA"
        )
        
        # Protein dosyasını kaydetme
        output_file_path = os.path.join(output_folder, f"{record.id}_protein.fasta")
        with open(output_file_path, "w") as output_file:
            SeqIO.write(protein_record, output_file, "fasta")
        
        print(f"Protein dizisi {output_file_path} dosyasına kaydedildi.")
```

## E-) DNA sıralarının GC içeriğinin hesaplanması

```python
# DNA sekans klasörü
input_folder = "DNA_sequence_10"

# Klasör yolunu oluştur
input_folder_path = os.path.join(os.getcwd(), input_folder)
files = os.listdir(input_folder_path)  # Klasördeki dosyaları listele

# GC içeriğini hesaplayan fonksiyon
def calculate_gc_content(sequence):
    g_count = sequence.count('G')
    c_count = sequence.count('C')
    gc_content = ((g_count + c_count) / len(sequence)) * 100
    return gc_content

# Her dosya için GC içeriğini hesaplama
for file in files:
    print(f"\n--- Dosya: {file} ---")
    file_path = os.path.join(input_folder_path, file)  # Dosyanın tam yolunu al
    
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        dna_sequence = str(record.seq)  # DNA dizisini stringe çevir
        gc_content = calculate_gc_content(dna_sequence)  # GC içeriğini hesapla
        
        print(f"ID: {record.id}")
        print(f"DNA Dizi Uzunluğu: {len(dna_sequence)}")
        print(f"GC İçeriği: {gc_content:.2f}%")
        print(f"\n*****************************************************************************************************")

```

## F-) DNA sıralarının Global Hizalarının Hesaplanması

```python
from Bio import SeqIO, Align

# DNA sekans klasörü
input_folder = "DNA_sequence_10"

# Klasör yolunu oluştur
input_folder_path = os.path.join(os.getcwd(), input_folder)
files = os.listdir(input_folder_path)  # Klasördeki dosyaları listele

# Global hizalama için PairwiseAligner nesnesi
aligner = Align.PairwiseAligner()
aligner.mode = "global"  # Global hizalama modu
aligner.match_score = 1.0  # Eşleşme puanı
aligner.mismatch_score = -1.0  # Uyumsuzluk cezası
aligner.open_gap_score = -2.0  # Açık boşluk cezası
aligner.extend_gap_score = -0.5  # Boşluk uzatma cezası

# DNA dizilerini okuma
sequences = []
sequence_ids = []

for file in files:
    file_path = os.path.join(input_folder_path, file)
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        sequences.append(str(record.seq))
        sequence_ids.append(record.id)

# Her iki dizi için global hizalama yapma
for i in range(len(sequences)):
    for j in range(i + 1, len(sequences)):
        seq1 = sequences[i]
        seq2 = sequences[j]
        alignment = aligner.align(seq1, seq2)[0]  # İlk hizalamayı al
        print("\n=================================================================================================")
        print(f"Hizalama: {sequence_ids[i]} vs {sequence_ids[j]}")
        print("=================================================================================================")
        print(alignment)
        print(f"Skor: {alignment.score}")
```

## G-) DNA sıralarının Lokal hizalarının hesaplanması

```python
# DNA sekans klasörü
input_folder = "DNA_sequence_10"

# Klasör yolunu oluştur
input_folder_path = os.path.join(os.getcwd(), input_folder)
files = os.listdir(input_folder_path)  # Klasördeki dosyaları listele

# Lokal hizalama için PairwiseAligner nesnesi
aligner = Align.PairwiseAligner()
aligner.mode = "local"  # Lokal hizalama modu
aligner.match_score = 1.0  # Eşleşme puanı
aligner.mismatch_score = -1.0  # Uyumsuzluk cezası
aligner.open_gap_score = -2.0  # Açık boşluk cezası
aligner.extend_gap_score = -0.5  # Boşluk uzatma cezası

# DNA dizilerini okuma
sequences = []
sequence_ids = []

for file in files:
    file_path = os.path.join(input_folder_path, file)
    for record in SeqIO.parse(file_path, "fasta"):  # FASTA formatında okuyun
        sequences.append(str(record.seq))
        sequence_ids.append(record.id)

# Her iki dizi için lokal hizalama yapma
for i in range(len(sequences)):
    for j in range(i + 1, len(sequences)):
        seq1 = sequences[i]
        seq2 = sequences[j]
        alignment = aligner.align(seq1, seq2)[0]  # İlk hizalamayı al
        print("\n==================================================================================================")
        print(f"Hizalama: {sequence_ids[i]} vs {sequence_ids[j]}")
        print("==================================================================================================")
        print(alignment)
        print(f"Skor: {alignment.score}")
```

## H-) Pozisyon Sıklık Matrislerinin (PFM) bulunması

```python
import numpy as np

# DNA dizilerini yükle
folder = "DNA_sequence_10"
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]

# DNA dizilerini listeye al
sequences = []

for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequences.append(str(record.seq))

# Maksimum uzunlukta tüm sekansları hizala (kısa olanları boşluklarla tamamla)
max_length = max(len(seq) for seq in sequences)
aligned_sequences = [seq.ljust(max_length, '-') for seq in sequences]  # '-' boşluğu temsil eder

# PFM hesaplama
nucleotides = ['A', 'T', 'C', 'G']
pfm = np.zeros((len(nucleotides), max_length))  # 4xMax_Length matris oluştur

for i, nucleotide in enumerate(nucleotides):
    for j in range(max_length):
        pfm[i, j] = sum(1 for seq in aligned_sequences if seq[j] == nucleotide)

# PFM çıktısı
print("\nPozisyon Sıklık Matrisi (PFM):")
for i, nucleotide in enumerate(nucleotides):
    print(f"{nucleotide}: {pfm[i]}")

# Daha okunaklı hale getirmek için tablo formatında
import pandas as pd

pfm_df = pd.DataFrame(pfm.T, columns=nucleotides)
pfm_df.index.name = "Pozisyon"
print("\nPozisyon Sıklık Matrisi (Tablo Formatında):")
pfm_df.head(10) #kaça kadar görülmek istenirse bu sayı ayarlanabilir.
```

## I-) Pozisyon Olasılık Matrislerinin (PPM) bulunması

```python
import numpy as np
import pandas as pd
from Bio import SeqIO

# DNA dizilerini yükle
folder = "DNA_sequence_10"
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]

# DNA dizilerini listeye al
sequences = []

for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequences.append(str(record.seq))

# Maksimum uzunlukta tüm sekansları hizala (kısa olanları boşluklarla tamamla)
max_length = max(len(seq) for seq in sequences)
aligned_sequences = [seq.ljust(max_length, '-') for seq in sequences]  # '-' boşluğu temsil eder

# PFM hesaplama
nucleotides = ['A', 'T', 'C', 'G']
pfm = np.zeros((len(nucleotides), max_length))  # 4xMax_Length matris oluştur

for i, nucleotide in enumerate(nucleotides):
    for j in range(max_length):
        pfm[i, j] = sum(1 for seq in aligned_sequences if seq[j] == nucleotide)


# PFM -> PPM Dönüşümü
sequence_count = len(sequences)
ppm = pfm / sequence_count  # Her pozisyondaki frekansı toplam sekans sayısına böl

# PPM çıktısı
print("\nPozisyon Olasılık Matrisi (PPM):")
for i, nucleotide in enumerate(nucleotides):
    print(f"{nucleotide}: {ppm[i]}")

# Daha okunaklı hale getirmek için tablo formatında
ppm_df = pd.DataFrame(ppm.T, columns=nucleotides)
ppm_df.index.name = "Pozisyon"
print("\nPozisyon Olasılık Matrisi (Tablo Formatında):")
ppm_df.head(10) #kaça kadar görülmek istenirse bu sayı ayarlanabilir.
```

## J-) Pozisyon Ağırlık Matrisinin (PWM) bulunması

```python
# Rastgele bir dağılım belirlenir (örneğin, nükleotidler eşit dağılımda ise her biri 0.25)
background_distribution = {'A': 0.25, 'T': 0.25, 'C': 0.25, 'G': 0.25}

# PWM hesaplama
def calculate_pwm(ppm_df, background_distribution):
    pwm = np.zeros_like(ppm_df)
    for nucleotide in ppm_df.columns:
        for position in ppm_df.index:
            probability = ppm_df.loc[position, nucleotide]
            if background_distribution[nucleotide] > 0:  # Bölme sıfırdan kaçınma
                pwm[position, ppm_df.columns.get_loc(nucleotide)] = np.log2(probability / background_distribution[nucleotide]) if probability > 0 else -np.inf
    return pwm

# PWM hesaplama
pwm_matrix = calculate_pwm(ppm_df, background_distribution)

# PWM matrisi çıktısı
pwm_df = pd.DataFrame(pwm_matrix, columns=ppm_df.columns, index=ppm_df.index)
pwm_df.index.name = "Pozisyon"
print("\nPWM Matrisi (Tablo Formatında):")
print(pwm_df.head(10))  # İlk 10 pozisyonu göster
```

# 2. BÖLÜM (PROTEİN)
## A-) Protein sıralarında 5 uzunluklu motiflerin bulunması

```python
import re
import random

## Bu kodda motif her turda random olarak seçilmek üzere ayarlanmıştır.

# Klasör adı
folder = "Protein_sequence_10"

# Rastgele 5 uzunluklu motif oluştur
amino_acids = "ACDEFGHIKLMNPQRSTVWY"  # Standart 20 amino asit
motif = "".join(random.choices(amino_acids, k=5))
# Burada belirli bir motif aranmak istenirse, motif değişkeni değiştirilebilir.
print(f"Aranacak rastgele motif: {motif}")

# Klasördeki dosyaları listele
file_paths = [os.path.join(folder, file) for file in os.listdir(folder)]

# Motif arama
print("\nProtein Sıralarında Motif Arama Sonuçları:\n" + "=" * 50)
for file_path in file_paths:
    for record in SeqIO.parse(file_path, "fasta"):
        sequence = str(record.seq)
        print(f"\n--- Dosya: {file_path} ---")
        print(f"ID: {record.id}")
        
        # Motif arama
        matches = list(re.finditer(motif, sequence))
        if matches:
            for match in matches:
                start = match.start()
                end = match.end()
                print(f"Motif bulundu: {match.group()} (Pozisyon: {start}-{end})")
        else:
            print(f"{motif} motifi bulunamadı.")
```
